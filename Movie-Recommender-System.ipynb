```python
# genres
# id
# keywords
# title
# overview
# cast
# crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
# ['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#movies['cast'][0]
```

```python
# we will take only first three actors 
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
             L.append(i['name'])
             counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
#movies['crew'][0]
```

```python
# we will only take director's name
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break 
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# removing space between names
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['genres']+movies['overview']+movies['cast']+movies['crew']+movies['keywords']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
import nltk
```

```python
# to convert ('love','loved','loving') -> ('love','love','love')
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
print(ps.stem('loved'))
print(ps.stem('loving'))
print(ps.stem('love'))
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))  # join list into string
new_df['tags'] = new_df['tags'].apply(stem) 
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

### Vectorization 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
cv.get_feature_names_out()
```

```python
print(cv.get_feature_names_out().tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# calculating distance between each vectors with each vectors
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
